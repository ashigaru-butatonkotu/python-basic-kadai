```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name} is an adult")
    else:
      print(f"{self.name} is not an adult")

human1 = Human("Takeshi", 33)
human2 = Human("Kim", 15)
human3 = Human("Ronaldo", 56)
human4 = Human("Kazuaki", 12)
human5 = Human("Kyoko", 18)
human6 = Human("Diana", 22)

humans = [human1,human2, human3, human4, human5, human6]

for human in humans:
  human.check_adult()
```

```python
class Human:
  def __init__(self, name, age, gender):
    self.name = name
    self.age = age
    self.gender = gender
  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name} ({self.gender})is an adult")
    else:
      print(f"{self.name} ({self.gender})is not an adult")

human1 = Human("Takeshi", 33, "male")
human2 = Human("Kim", 15, "female")
human3 = Human("Ronaldo", 56, "male")
human4 = Human("Kazuaki", 12,"male")
human5 = Human("Kyoko", 18, "female")
human6 = Human("Diana", 22, "male")

humans = [human1,human2, human3, human4, human5, human6]

for human in humans:
  human.check_adult()
```
